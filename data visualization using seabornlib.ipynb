```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =sns.load_dataset("tips")
```

```python
df
```

# univariant "one variable" data visualization

```python
sns.stripplot(y="tip", data=df,jitter=True)
plt.ylabel("tip ($)")
plt.show()
```

```python
plt.figure(figsize=(8,7))
sns.boxplot(x="size", y="tip", data=df,)
plt.ylabel("tip ($)")
plt.show()
```

```python
plt.figure(figsize=(8,7))
sns.boxplot(x="time", y="tip", data=df,)
plt.ylabel("tip ($)")
plt.show()
```

```python
sns.boxplot(x="day", y="tip", hue="sex", data=df)
plt.ylabel("tip ($)")
plt.show()
```

```python
sns.stripplot(x="day", y="tip",hue="sex", data=df,jitter=True)
plt.ylabel("tip ($)")
plt.show()
```

```python
plt.figure(figsize=(8,7))
sns.boxplot(x="smoker", y="tip", data=df,)
plt.ylabel("tip ($)")
plt.show()
```

```python
sns.violinplot(x="sex",y="total_bill",data=df)
plt.title("comparing male and female paying total bill")
plt.show()
```
